# 99 — Full Pipeline Dashboard

End-to-end presentation notebook: data -> statistics -> forecasting -> signals -> portfolio -> advanced comparison.

```python
from pathlib import Path
import json
import pandas as pd
from IPython.display import display, Image, Markdown

ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
TABLES = ROOT / 'project report' / 'tables'
GRAPHICS = ROOT / 'project report' / 'graphics'
PROCESSED = ROOT / 'data' / 'processed'
MANIFEST = PROCESSED / 'demo_run_manifest.json'

if MANIFEST.exists():
    manifest = json.loads(MANIFEST.read_text(encoding='utf-8'))
    display(Markdown('## Run Manifest'))
    base_manifest = {k: v for k, v in manifest.items() if k != 'notebook_readiness'}
    display(pd.DataFrame([base_manifest]))
    if 'notebook_readiness' in manifest:
        display(Markdown('## Notebook Readiness'))
        display(pd.DataFrame(manifest['notebook_readiness']))
else:
    print('Manifest not found. Run 00_cache_and_run_control.ipynb first.')
```

## Data and Statistical Evidence

```python
for name in [
    'fig_4_2_russian_rates.png',
    'fig_4_3_lpr_fx.png',
    'fig_4_5_chinese_bonds.png',
    'fig_4_7_global.png',
    'fig_5_1_correlation.png',
    'fig_5_2_irf.png',
]:
    p = GRAPHICS / name
    display(Markdown(f'### {name}'))
    if p.exists():
        display(Image(filename=str(p)))
    else:
        print(f'Missing: {p}')
```

## Forecasting and OOS Comparison

```python
ru = pd.read_csv(TABLES / 'tab_6_3_model_comparison_ru.csv', index_col=0)
cn = pd.read_csv(TABLES / 'tab_6_3_model_comparison_cn.csv', index_col=0)
best = pd.read_csv(TABLES / 'tab_6_3_model_comparison_best_model.csv')

display(Markdown('### RU OOS RMSE'))
display(ru)
display(Markdown('### CN OOS RMSE'))
display(cn)
display(Markdown('### Best Model by Maturity'))
display(best)

avg_img = TABLES / 'tab_6_3_model_comparison_avg_rmse.png'
if avg_img.exists():
    display(Image(filename=str(avg_img)))
```

## Signals and Portfolio Outcomes

```python
def safe_csv(path):
    if path.exists():
        return pd.read_csv(path)
    return pd.DataFrame()

snapshot = safe_csv(PROCESSED / 'signals_latest_snapshot.csv')
diag = safe_csv(PROCESSED / 'signals_latest_diagnostics.csv')
pf = safe_csv(PROCESSED / 'portfolio_backtest_summary.csv')
ml_snapshot = safe_csv(PROCESSED / 'ml_signals_snapshot.csv')
ml_diag = safe_csv(PROCESSED / 'ml_signals_diagnostics.csv')
ml_pf = safe_csv(PROCESSED / 'ml_backtest_summary.csv')
adv = safe_csv(TABLES / 'tab_6_4_advanced_methods.csv')

display(Markdown('### Latest Signal Snapshot (Classical)'))
display(snapshot)
display(Markdown('### Latest Signal Snapshot (ML)'))
display(ml_snapshot)
display(Markdown('### Signal Stability Diagnostics (Classical)'))
display(diag)
display(Markdown('### ML Diagnostics'))
display(ml_diag)
display(Markdown('### Portfolio Backtest Summary (Classical)'))
display(pf)
display(Markdown('### Portfolio Backtest Summary (ML)'))
display(ml_pf)
display(Markdown('### Advanced Method Diagnostics'))
display(adv)
```
