```python
"""
============================================================================
DATABASE CREATION: Russian & Chinese Bond Yields + Macroeconomic Factors
============================================================================
Data Sources:
- Russian Bond Yields: Tinkoff/Alfa/BCS Broker APIs (API keys required)
- Chinese Bond Yields: People's Bank of China / FRED
- Macroeconomic Factors: World Bank, FRED, Trading Economics APIs
============================================================================
"""
import pandas as pd
import numpy as np
import requests
import sqlite3
from io import StringIO
from datetime import datetime, timedelta
import xml.etree.ElementTree as ET
import warnings
warnings.filterwarnings('ignore')

DB_PATH = "bond_rates_database.db"

# ==========================================================================
# API KEYS CONFIGURATION (Add your keys here when available)
# ==========================================================================
# 
# INSTRUCTIONS:
# 1. Get API keys from:
#    - Tinkoff: https://www.tinkoff.ru/invest/open-api/ (Invest API token)
#    - Alfa: Contact Alfa-Bank for API access
#    - BCS: Contact BCS for API access
#    - FRED: https://fred.stlouisfed.org/docs/api/api_key.html (free)
#
# 2. Add your keys below (replace None with your actual keys):
#
API_KEYS = {
    'tinkoff': None,  # Tinkoff Invest API token (Bearer token)
    'alfa': None,     # Alfa-Bank API key
    'bcs': None,      # BCS API key
    'fred': None,     # FRED API key (optional, free tier available)
    'tradingeconomics': None,  # Trading Economics API key (optional)
}

# Example (uncomment and fill when you have keys):
# API_KEYS = {
#     'tinkoff': 'your_tinkoff_token_here',
#     'alfa': 'your_alfa_key_here',
#     'bcs': 'your_bcs_key_here',
#     'fred': 'your_fred_key_here',  # Optional
#     'tradingeconomics': None,
# }

# ==========================================================================
# PART 1: RUSSIAN BOND YIELDS FROM BROKER APIs
# ==========================================================================

def fetch_tinkoff_bond_yields(api_token: str, date_from: str, date_to: str) -> pd.DataFrame:
    """
    Fetch Russian bond yields from Tinkoff Invest API.
    Requires: Tinkoff Invest API token
    Docs: https://www.tinkoff.ru/invest/open-api/
    """
    if not api_token:
        return pd.DataFrame()
    
    base_url = "https://invest-public-api.tinkoff.ru/rest/"
    headers = {
        "Authorization": f"Bearer {api_token}",
        "Content-Type": "application/json"
    }
    
    # Tinkoff API endpoints for bonds
    # Note: This is a placeholder structure - actual implementation depends on Tinkoff API version
    try:
        # Get OFZ bonds list
        bonds_url = f"{base_url}tinkoff.public.invest.api.contract.v1.InstrumentsService/Bonds"
        response = requests.post(bonds_url, headers=headers, json={
            "instrument_status": "INSTRUMENT_STATUS_BASE"
        }, timeout=30)
        
        if response.status_code == 200:
            data = response.json()
            # Process bond data
            # Implementation depends on actual API response structure
            print("Tinkoff API: Connected successfully")
            # TODO: Implement actual bond yield fetching logic
            return pd.DataFrame()
    except Exception as e:
        print(f"Tinkoff API error: {e}")
    
    return pd.DataFrame()

def fetch_alfa_bond_yields(api_key: str, date_from: str, date_to: str) -> pd.DataFrame:
    """
    Fetch Russian bond yields from Alfa-Bank API.
    Requires: Alfa-Bank API key
    """
    if not api_key:
        return pd.DataFrame()
    
    # Alfa-Bank API endpoint (placeholder - actual endpoint may vary)
    base_url = "https://api.alfabank.ru/"
    headers = {
        "X-API-Key": api_key,
        "Content-Type": "application/json"
    }
    
    try:
        # Placeholder for Alfa API implementation
        print("Alfa API: Structure ready (implementation pending API key)")
        return pd.DataFrame()
    except Exception as e:
        print(f"Alfa API error: {e}")
    
    return pd.DataFrame()

def fetch_bcs_bond_yields(api_key: str, date_from: str, date_to: str) -> pd.DataFrame:
    """
    Fetch Russian bond yields from BCS API.
    Requires: BCS API key
    """
    if not api_key:
        return pd.DataFrame()
    
    # BCS API endpoint (placeholder - actual endpoint may vary)
    base_url = "https://api.bcs.ru/"
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }
    
    try:
        # Placeholder for BCS API implementation
        print("BCS API: Structure ready (implementation pending API key)")
        return pd.DataFrame()
    except Exception as e:
        print(f"BCS API error: {e}")
    
    return pd.DataFrame()

# Try broker APIs in order of preference
print("Fetching Russian bond yields from broker APIs...")
ru_yields_broker = []

# Try Tinkoff
if API_KEYS['tinkoff']:
    print("  Trying Tinkoff API...")
    df = fetch_tinkoff_bond_yields(API_KEYS['tinkoff'], "2015-01-01", datetime.now().strftime("%Y-%m-%d"))
    if not df.empty:
        ru_yields_broker.append(df)

# Try Alfa
if API_KEYS['alfa']:
    print("  Trying Alfa API...")
    df = fetch_alfa_bond_yields(API_KEYS['alfa'], "2015-01-01", datetime.now().strftime("%Y-%m-%d"))
    if not df.empty:
        ru_yields_broker.append(df)

# Try BCS
if API_KEYS['bcs']:
    print("  Trying BCS API...")
    df = fetch_bcs_bond_yields(API_KEYS['bcs'], "2015-01-01", datetime.now().strftime("%Y-%m-%d"))
    if not df.empty:
        ru_yields_broker.append(df)

if not ru_yields_broker:
    print("  ⚠ No broker API keys provided. Please add API keys to API_KEYS dictionary.")
    print("  Available brokers: Tinkoff, Alfa, BCS")
    ru_wide = pd.DataFrame(columns=["date"])
else:
    # Process broker data
    ru_yields = pd.concat(ru_yields_broker, ignore_index=True)
    # Resample to monthly
    ru_yields["month"] = ru_yields["date"].dt.to_period("M")
    ru_monthly = ru_yields.groupby(["month", "maturity"]).agg({"yield": "mean"}).reset_index()
    ru_monthly["date"] = ru_monthly["month"].dt.to_timestamp()
    ru_monthly = ru_monthly.drop(columns=["month"])
    
    # Pivot to wide format
    ru_wide = ru_monthly.pivot(index="date", columns="maturity", values="yield")
    ru_wide.columns = [f"RU_{col}" for col in ru_wide.columns]
    ru_wide = ru_wide.reset_index()
    print(f"  ✓ Russian yields from broker APIs: {len(ru_wide)} monthly observations")

# Fallback: CBR yield curve (if broker APIs fail)
def fetch_cbr_yield_curve(date_from: str, date_to: str) -> pd.DataFrame:
    """
    Fetch Russian government bond yield curve from CBR.
    Uses the G-Curve (zero-coupon yield curve) parameters.
    """
    # CBR provides yield curve parameters, we calculate yields for specific maturities
    url = "https://cbr.ru/hd_base/zcyc_params/zcyc/"
    
    # Try to get the page with parameters
    try:
        params = {
            "UniDbQuery.Posted": "True",
            "UniDbQuery.From": date_from,
            "UniDbQuery.To": date_to,
        }
        r = requests.get(url, params=params, timeout=30)
        r.raise_for_status()
        
        # Parse HTML table (CBR provides data in HTML format)
        tables = pd.read_html(StringIO(r.text), decimal=',', thousands=' ')
        if tables:
            df = tables[0]
            return df
    except Exception as e:
        print(f"CBR direct fetch failed: {e}")
    
    return pd.DataFrame()

def calculate_nelson_siegel_yield(t, beta0, beta1, beta2, tau):
    """Calculate yield using Nelson-Siegel model parameters."""
    if tau == 0:
        return np.nan
    factor = (1 - np.exp(-t/tau)) / (t/tau)
    return beta0 + beta1 * factor + beta2 * (factor - np.exp(-t/tau))

# Alternative: Use MOEX historical data for OFZ bonds
def fetch_moex_ofz_history(secid: str, date_from: str, date_to: str) -> pd.DataFrame:
    """Fetch historical yield data for a specific OFZ bond from MOEX."""
    url = f"https://iss.moex.com/iss/history/engines/stock/markets/bonds/securities/{secid}.json"
    all_data = []
    start = 0
    
    while True:
        params = {"from": date_from, "till": date_to, "start": start, "limit": 100}
        try:
            r = requests.get(url, params=params, timeout=30)
            r.raise_for_status()
            payload = r.json()
            
            if "history" not in payload:
                break
                
            cols = payload["history"]["columns"]
            rows = payload["history"]["data"]
            
            if not rows:
                break
                
            df = pd.DataFrame(rows, columns=cols)
            all_data.append(df)
            start += 100
            
            if len(rows) < 100:
                break
        except Exception as e:
            print(f"MOEX fetch error for {secid}: {e}")
            break
    
    if all_data:
        return pd.concat(all_data, ignore_index=True)
    return pd.DataFrame()

# List of representative OFZ bonds for different maturities
# These are commonly used benchmark bonds
OFZ_BONDS = {
    "1Y": ["SU26234RMFS1", "SU26227RMFS5"],  # Short-term
    "2Y": ["SU26229RMFS1", "SU26233RMFS3"],
    "3Y": ["SU26232RMFS5", "SU26230RMFS9"],
    "5Y": ["SU26235RMFS9", "SU26236RMFS7"],
    "10Y": ["SU26238RMFS3", "SU26240RMFS9", "SU26241RMFS7"],
}

print("Fetching Russian bond data from MOEX...")

# Fetch historical data for Russian bonds
ru_yields_monthly = []
date_from = "2015-01-01"
date_to = datetime.now().strftime("%Y-%m-%d")

for maturity, secids in OFZ_BONDS.items():
    for secid in secids:
        df = fetch_moex_ofz_history(secid, date_from, date_to)
        if not df.empty and "TRADEDATE" in df.columns:
            # Look for yield column
            yield_col = None
            for col in ["YIELDCLOSE", "YIELD", "YIELDLASTCOUPON", "YIELDATWAPRICE"]:
                if col in df.columns:
                    yield_col = col
                    break
            
            if yield_col:
                df["date"] = pd.to_datetime(df["TRADEDATE"])
                df["yield"] = pd.to_numeric(df[yield_col], errors="coerce")
                df["maturity"] = maturity
                df["secid"] = secid
                df = df[["date", "yield", "maturity", "secid"]].dropna()
                if not df.empty:
                    ru_yields_monthly.append(df)
                    print(f"  Fetched {len(df)} records for {secid} ({maturity})")
                    break  # Got data for this maturity, move on

if ru_yields_monthly:
    ru_yields = pd.concat(ru_yields_monthly, ignore_index=True)
    # Resample to monthly
    ru_yields["month"] = ru_yields["date"].dt.to_period("M")
    ru_monthly = ru_yields.groupby(["month", "maturity"]).agg({"yield": "mean"}).reset_index()
    ru_monthly["date"] = ru_monthly["month"].dt.to_timestamp()
    ru_monthly = ru_monthly.drop(columns=["month"])
    
    # Pivot to wide format
    ru_wide = ru_monthly.pivot(index="date", columns="maturity", values="yield")
    ru_wide.columns = [f"RU_{col}" for col in ru_wide.columns]
    ru_wide = ru_wide.reset_index()
    print(f"Russian yields: {len(ru_wide)} monthly observations")
else:
    ru_wide = pd.DataFrame(columns=["date"])
    print("No Russian yield data fetched")

ru_wide.head()
```

```python
# ==========================================================================
# PART 3: MACROECONOMIC FACTORS FROM ALTERNATIVE SOURCES (Monthly Data)
# ==========================================================================
print("Fetching macroeconomic data from alternative sources...")

# ==========================================================================
# WORLD BANK API - Monthly/Quarterly Indicators
# ==========================================================================
def fetch_world_bank_indicator(indicator: str, country_code: str, date_range: str = "2010:2025") -> pd.DataFrame:
    """
    Fetch indicator data from World Bank API.
    Country codes: RUS (Russia), CHN (China)
    """
    url = f"https://api.worldbank.org/v2/country/{country_code}/indicator/{indicator}"
    params = {"format": "json", "date": date_range, "per_page": 500}
    try:
        r = requests.get(url, params=params, timeout=30)
        r.raise_for_status()
        data = r.json()
        if len(data) > 1 and data[1]:
            records = []
            for d in data[1]:
                if d.get("value") is not None:
                    records.append({
                        "date": pd.to_datetime(f"{d['date']}-01-01"),  # Annual data
                        "value": d["value"]
                    })
            if records:
                df = pd.DataFrame(records)
                df = df.sort_values("date")
                return df
    except Exception as e:
        print(f"  World Bank fetch failed for {indicator}: {e}")
    return pd.DataFrame()

# ==========================================================================
# FRED API - Federal Reserve Economic Data (Monthly)
# ==========================================================================
def fetch_fred_series(series_id: str, api_key: str = None) -> pd.DataFrame:
    """
    Fetch data from FRED API.
    No API key needed for basic CSV downloads, but key enables more requests.
    """
    # Method 1: CSV download (no key needed)
    csv_url = f"https://fred.stlouisfed.org/graph/fredgraph.csv?id={series_id}"
    try:
        df = pd.read_csv(csv_url, parse_dates=["DATE"])
        df.columns = ["date", "value"]
        df = df.dropna()
        return df
    except Exception as e:
        print(f"  FRED CSV download failed for {series_id}: {e}")
    
    # Method 2: JSON API (requires key for higher limits)
    if api_key:
        api_url = f"https://api.stlouisfed.org/fred/series/observations"
        params = {
            "series_id": series_id,
            "api_key": api_key,
            "file_type": "json",
            "frequency": "m",  # Monthly
            "sort_order": "asc"
        }
        try:
            r = requests.get(api_url, params=params, timeout=30)
            r.raise_for_status()
            data = r.json()
            if "observations" in data:
                records = [{"date": pd.to_datetime(obs["date"]), "value": float(obs["value"])} 
                          for obs in data["observations"] if obs["value"] != "."]
                return pd.DataFrame(records)
        except Exception as e:
            print(f"  FRED API error for {series_id}: {e}")
    
    return pd.DataFrame()

# ==========================================================================
# RUSSIAN MACROECONOMIC INDICATORS
# ==========================================================================
print("\nFetching Russian macroeconomic indicators...")

# World Bank indicators for Russia
wb_ru_indicators = {
    'NY.GDP.MKTP.KD.ZG': 'gdp_growth',           # GDP growth (annual %)
    'FP.CPI.TOTL.ZG': 'inflation_cpi',          # Inflation, consumer prices (annual %)
    'SL.UEM.TOTL.ZS': 'unemployment',            # Unemployment, total (% of total labor force)
    'NY.GDP.MKTP.CD': 'gdp_usd_bn',             # GDP (current US$)
    'GC.DOD.TOTL.GD.ZS': 'govt_debt_pct',       # Central government debt, total (% of GDP)
    'NE.TRD.GNFS.ZS': 'trade_pct',              # Trade (% of GDP)
    'FR.INR.LEND': 'lending_rate',               # Lending interest rate (%)
}

ru_macro_frames = []
for indicator_code, name in wb_ru_indicators.items():
    df = fetch_world_bank_indicator(indicator_code, "RUS", "2010:2025")
    if not df.empty:
        df.columns = ['date', f'RU_{name}']
        ru_macro_frames.append(df)
        print(f"  ✓ {name}: {len(df)} records")

# FRED series for Russia (if available)
fred_ru_series = {
    # Add FRED series IDs for Russian data if available
}

for series_id, name in fred_ru_series.items():
    df = fetch_fred_series(series_id, API_KEYS.get('fred'))
    if not df.empty:
        df.columns = ['date', f'RU_{name}']
        ru_macro_frames.append(df)
        print(f"  ✓ {name} (FRED): {len(df)} records")

# Combine Russian macro data
if ru_macro_frames:
    ru_macro = ru_macro_frames[0]
    for df in ru_macro_frames[1:]:
        ru_macro = ru_macro.merge(df, on='date', how='outer')
    ru_macro = ru_macro.sort_values('date')
    # Convert annual to monthly (forward fill)
    ru_macro['year_month'] = ru_macro['date'].dt.to_period('M')
    ru_macro_monthly = ru_macro.set_index('date').resample('ME').ffill().reset_index()
    ru_macro_monthly = ru_macro_monthly.drop(columns=['year_month'], errors='ignore')
    print(f"\n  Russian macro: {len(ru_macro_monthly)} monthly observations")
else:
    ru_macro_monthly = pd.DataFrame(columns=['date'])
    print("  ⚠ No Russian macro data fetched")

# ==========================================================================
# CHINESE MACROECONOMIC INDICATORS
# ==========================================================================
print("\nFetching Chinese macroeconomic indicators...")

# World Bank indicators for China
wb_cn_indicators = {
    'NY.GDP.MKTP.KD.ZG': 'gdp_growth',
    'FP.CPI.TOTL.ZG': 'inflation_cpi',
    'SL.UEM.TOTL.ZS': 'unemployment',
    'NY.GDP.MKTP.CD': 'gdp_usd_bn',
    'GC.DOD.TOTL.GD.ZS': 'govt_debt_pct',
    'NE.TRD.GNFS.ZS': 'trade_pct',
    'FR.INR.LEND': 'lending_rate',
}

cn_macro_frames = []
for indicator_code, name in wb_cn_indicators.items():
    df = fetch_world_bank_indicator(indicator_code, "CHN", "2010:2025")
    if not df.empty:
        df.columns = ['date', f'CN_{name}']
        cn_macro_frames.append(df)
        print(f"  ✓ {name}: {len(df)} records")

# FRED series for China
fred_cn_series = {
    'NAEXKP01CNQ189S': 'gdp_growth',  # Real GDP per Capita (if available)
}

for series_id, name in fred_cn_series.items():
    df = fetch_fred_series(series_id, API_KEYS.get('fred'))
    if not df.empty:
        df.columns = ['date', f'CN_{name}']
        cn_macro_frames.append(df)
        print(f"  ✓ {name} (FRED): {len(df)} records")

# Combine Chinese macro data
if cn_macro_frames:
    cn_macro = cn_macro_frames[0]
    for df in cn_macro_frames[1:]:
        cn_macro = cn_macro.merge(df, on='date', how='outer')
    cn_macro = cn_macro.sort_values('date')
    # Convert annual to monthly (forward fill)
    cn_macro['year_month'] = cn_macro['date'].dt.to_period('M')
    cn_macro_monthly = cn_macro.set_index('date').resample('ME').ffill().reset_index()
    cn_macro_monthly = cn_macro_monthly.drop(columns=['year_month'], errors='ignore')
    print(f"\n  Chinese macro: {len(cn_macro_monthly)} monthly observations")
else:
    cn_macro_monthly = pd.DataFrame(columns=['date'])
    print("  ⚠ No Chinese macro data fetched")

# Display samples
print("\n" + "-"*70)
print("Russian macro columns:", list(ru_macro_monthly.columns) if not ru_macro_monthly.empty else "None")
print("Chinese macro columns:", list(cn_macro_monthly.columns) if not cn_macro_monthly.empty else "None")
print("-"*70)

ru_macro_monthly.tail() if not ru_macro_monthly.empty else None, cn_macro_monthly.tail() if not cn_macro_monthly.empty else None
```

```python
import requests
import pandas as pd
from sqlalchemy import create_engine

def fetch_imf_weo_gdp():
    url = (
        "https://dataservices.imf.org/REST/SDMX_JSON.svc/"
        "CompactData/WEO/Q.CN.NGDPD"
    )

    r = safe_get(url)
    data = r.json()

    series = data["CompactData"]["DataSet"]["Series"]["Obs"]

    rows = []
    for obs in series:
        rows.append({
            "date": pd.to_datetime(obs["@TIME_PERIOD"]),
            "gdp_quarterly": float(obs["@OBS_VALUE"])
        })

    return pd.DataFrame(rows)

def expand_quarterly_to_monthly(df_q):
    """
    Split quarterly GDP equally across the 3 months of the quarter
    """
    monthly_rows = []

    for _, row in df_q.iterrows():
        quarter_end = row["date"]
        quarterly_value = row["gdp_quarterly"]
        monthly_value = quarterly_value / 3

        # Generate the 3 months in the quarter
        months = pd.date_range(
            end=quarter_end,
            periods=3,
            freq="M"
        )

        for m in months:
            monthly_rows.append({
                "country": "China",
                "indicator_code": "GDP_NOMINAL_MONTHLY_EQ",
                "indicator_name": "Nominal GDP (Quarterly split equally across months)",
                "date": m,
                "value": monthly_value,
                "unit": "Local currency",
                "source": "IMF IFS (quarterly, mechanically split)"
            })

    return pd.DataFrame(monthly_rows)

gdp_q = fetch_imf_quarterly_gdp()
gdp_monthly = expand_quarterly_to_monthly(gdp_q)

print(gdp_monthly.tail(6))

# TABLE macro_monthly (
#     country TEXT,
#     indicator_code TEXT,
#     indicator_name TEXT,
#     date DATE,
#     value FLOAT,
#     unit TEXT,
#     source TEXT
# )

def fetch_worldbank_monthly(indicator, indicator_name):
    """
    Fetch monthly China macro data from World Bank API
    """
    url = f"https://api.worldbank.org/v2/country/CHN/indicator/{indicator}"
    params = {
        "format": "json",
        "per_page": 20000
    }

    r = requests.get(url, params=params)
    data = r.json()[1]

    rows = []
    for d in data:
        if d["value"] is not None:
            rows.append({
                "country": "China",
                "indicator_code": indicator,
                "indicator_name": indicator_name,
                "date": pd.to_datetime(d["date"] + "-01"),
                "value": float(d["value"]),
                "unit": d["indicator"]["value"],
                "source": "World Bank"
            })

    return pd.DataFrame(rows)


# Example: CPI (monthly)
cpi_df = fetch_worldbank_monthly(
    indicator="FP.CPI.TOTL",
    indicator_name="Consumer Price Index"
)

print(cpi_df.tail())

def fetch_imf_ifs(series_code, indicator_name, unit):
    """
    Fetch IMF IFS monthly data for China
    """
    url = (
        "https://dataservices.imf.org/REST/SDMX_JSON.svc/"
        f"CompactData/IFS/M.CN.{series_code}"
    )

    r = requests.get(url)
    data = r.json()

    series = data["CompactData"]["DataSet"]["Series"]["Obs"]

    rows = []
    for obs in series:
        rows.append({
            "country": "China",
            "indicator_code": series_code,
            "indicator_name": indicator_name,
            "date": pd.to_datetime(obs["@TIME_PERIOD"]),
            "value": float(obs["@OBS_VALUE"]),
            "unit": unit,
            "source": "IMF – IFS"
        })

    return pd.DataFrame(rows)


# Examples
cpi_imf = fetch_imf_ifs("PCPI_IX", "Consumer Price Index", "Index")
ppi_imf = fetch_imf_ifs("PPI_IX", "Producer Price Index", "Index")
fx_imf  = fetch_imf_ifs("ENDA_XDC_USD_RATE", "CNY/USD Exchange Rate", "CNY per USD")

print(cpi_imf.tail())

macro_monthly.sort_values("date", inplace=True)

macro_monthly = pd.concat([
    gdp_monthly,
    cpi_df,
    ppi_df,
    fx_df
], ignore_index=True)

macro_monthly.sort_values("date", inplace=True)

engine = create_engine("sqlite:///china_macro_monthly.db")

macro_monthly.to_sql(
    "macro_monthly",
    engine,
    if_exists="replace",
    index=False
)

print("Monthly China macro database with GDP created.")
```

```python
import pandas as pd
from pathlib import Path

imf_path = Path("IMF_dataset_1.csv")
bond_path = Path("government_bond_index_monthly.csv")

bond_raw = pd.read_csv(bond_path, sep=';', decimal=',', encoding='cp1251', skiprows=1)
# ensure expected column names
bond_raw.columns = [c.strip().upper() for c in bond_raw.columns]

bond_raw['TRADEDATE'] = pd.to_datetime(bond_raw['TRADEDATE'], dayfirst=True, errors='coerce')
bond_raw['CLOSE'] = pd.to_numeric(bond_raw['CLOSE'], errors='coerce')

bond = bond_raw[['TRADEDATE', 'CLOSE']].dropna()

bond_monthly = bond.set_index('TRADEDATE').resample('M').mean().reset_index()
bond_annual = bond.set_index('TRADEDATE').resample('A').mean().reset_index()
bond_annual['year'] = bond_annual['TRADEDATE'].dt.year
bond_annual = bond_annual[['year', 'CLOSE']]

bond_monthly.head(), bond_annual.tail()
```

```python
imf = pd.read_csv(imf_path)
year_cols_all = [c for c in imf.columns if c.isdigit()]

def build_macro_wide(country_pattern, min_year, min_non_nan=5, name_cut=40):
    subset = imf[imf['COUNTRY'].str.contains(country_pattern, na=False)].copy()
    tidy = subset.melt(
        id_vars=['SERIES_CODE', 'INDICATOR'],
        value_vars=year_cols_all,
        var_name='year',
        value_name='value'
    )
    tidy['year'] = tidy['year'].astype(int)
    tidy['value'] = pd.to_numeric(tidy['value'], errors='coerce')
    tidy = tidy.dropna(subset=['value'])
    tidy = tidy[tidy['year'] >= min_year]

    wide_local = tidy.pivot_table(index='year', columns='SERIES_CODE', values='value')
    wide_local = wide_local.loc[:, wide_local.notna().sum() >= min_non_nan]

    ind_map = subset[['SERIES_CODE', 'INDICATOR']].drop_duplicates().set_index('SERIES_CODE')['INDICATOR']
    rename_map = {}
    for code, txt in ind_map.items():
        if not isinstance(txt, str):
            continue
        short = txt.split(',')[0].strip()[:name_cut]
        rename_map[code] = f"{code}_{short}"
    wide_local = wide_local.rename(columns=rename_map)
    return wide_local

wide_ru = build_macro_wide('Russian Federation', bond_annual['year'].min())
wide_ru.head()
```

```python
annual = bond_annual.merge(wide_ru, left_on='year', right_index=True, how='left')

annual_next = annual.assign(next_close=annual['CLOSE'].shift(-1)).dropna(subset=['next_close'])

corr_matrix = annual_next.corr(numeric_only=True)

corr = corr_matrix['next_close'].dropna().sort_values(ascending=False)
# display(corr.head(12), corr.tail(5))

# corr_matrix
```

```python
wide_cn = build_macro_wide('China', bond_annual['year'].min())

annual_cn = bond_annual.merge(wide_cn, left_on='year', right_index=True, how='left')
annual_cn_next = annual_cn.assign(next_close=annual_cn['CLOSE'].shift(-1)).dropna(subset=['next_close'])

corr_matrix_cn = annual_cn_next.corr(numeric_only=True)

corr_cn = corr_matrix_cn['next_close'].dropna().sort_values(ascending=False)
# display(corr_cn.head(12), corr_cn.tail(5))

# corr_matrix_cn
corr
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

model_data = annual.dropna()
X = model_data.drop(columns=['CLOSE', 'year'])
y = model_data['CLOSE']

if len(model_data) >= 8:
    split_year = model_data['year'].max() - 2
else:
    split_year = model_data['year'].max()

train_idx = model_data['year'] < split_year
X_train, X_test = X[train_idx], X[~train_idx]
y_train, y_test = y[train_idx], y[~train_idx]

linreg = make_pipeline(StandardScaler(), LinearRegression())
linreg.fit(X_train, y_train)
pred_lr = linreg.predict(X_test)

rf = RandomForestRegressor(n_estimators=400, random_state=42)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

def eval_model(name, y_true, y_pred):
    return {
        'model': name,
        'rmse': mean_squared_error(y_true, y_pred, squared=False),
        'r2': r2_score(y_true, y_pred)
    }

results = pd.DataFrame([
    eval_model('LinearRegression', y_test, pred_lr),
    eval_model('RandomForest', y_test, pred_rf)
])
display(results)

plot_df = pd.DataFrame({
    'year': model_data['year'][~train_idx],
    'actual': y_test.values,
    'linreg_pred': pred_lr,
    'rf_pred': pred_rf
})

plt.figure(figsize=(8, 4))
plt.plot(plot_df['year'], plot_df['actual'], label='Actual', marker='o')
plt.plot(plot_df['year'], plot_df['linreg_pred'], label='LinearReg', marker='o')
plt.plot(plot_df['year'], plot_df['rf_pred'], label='RandomForest', marker='o')
plt.xlabel('Year')
plt.ylabel('RGBI close (avg)')
plt.title('Bond index predictions from macro factors')
plt.legend()
plt.grid(True)
plt.show()
```
